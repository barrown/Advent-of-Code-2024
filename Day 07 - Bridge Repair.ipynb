```python
operator_combos = []
for op1 in ['+','*','|']:
    for op2 in ['+','*','|']:
        for op3 in ['+','*','|']:
            for op4 in ['+','*','|']:
                for op5 in ['+','*','|']:
                    for op6 in ['+','*','|']:
                        for op7 in ['+','*','|']:
                            for op8 in ['+','*','|']:
                                for op9 in ['+','*','|']:
                                    for op10 in ['+','*','|']:
                                        for op11 in ['+','*','|']:
                                            for op12 in ['+','*','|']:
                                                operator_combos.append(''.join([op1,op2,op3,op4,op5,op6,op7,op8,op9,op10,op11,op12]))
print(len(operator_combos))
```

```python
def return_subset(length):
    subset = set()
    for operator in operator_combos:
        subset.add(operator[:length])
    return subset
```

```python
dict_of_subsets = {}
for i in range(1,13):
    dict_of_subsets[i] = return_subset(i)
```

```python
with open("day_07_input_0.txt") as fname:
    test_input = fname.readlines()
```

```python
equations = []
for line in test_input:
    splitline = line.split(":")
    equations.append([splitline[0],splitline[1].strip()])
```

```python
for eq in equations:
    answer = int(eq[0])
    elements = eq[1].split()

    for operator in dict_of_subsets[len(elements)]:
        print("\n", eq, operator)
        construct = '('*len(elements)
        for i, number in enumerate(elements):
            if operator[i] == '|':
                print(construct)
                construct = construct[:-1] + number + ')' # take off the preceeding bracket and just paste the number in
                print(construct)
            else:
                construct += number + ')' + operator[i]

        print(operator.count('|'))
        truncate = operator.count('|') - 1
        print((construct))
        print(construct[truncate:])

        if truncate >= 0:
            if eval(construct[truncate:]) == answer:
                print(construct[truncate:],"=",eval(construct[truncate:]))
                accumulator += answer
                break
        else:
            if eval(construct) == answer:
                print(construct,"=",eval(construct))
                accumulator += answer
                break
```

```python
data = []
with open("day_07_input_1.txt", 'r') as f:
    data = f.read().strip('\n').split('\n')

def can_make(result, rest):
    if len(rest) == 1:
        return rest[0] == result

    last = rest[-1]

    if result % last == 0:
        possible_mul = can_make(result // last, rest[:-1])
    else:
        possible_mul = False

    next_power_of_10 = 1
    while next_power_of_10 <= last:
        next_power_of_10 *= 10
    if (result - last) % next_power_of_10 == 0:
        possible_concat = can_make((result - last) // next_power_of_10, rest[:-1])
    else:
        possible_concat = False

    possible_add = can_make(result - last, rest[:-1])
    return possible_mul or possible_add or possible_concat

total = 0
for line in data:
    result, rest = line.split(': ')
    result = int(result)
    rest = [int(x) for x in rest.split()]
    if can_make(result, rest):
        total += result

print(total)
```
