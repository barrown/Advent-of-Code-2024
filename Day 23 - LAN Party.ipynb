```python
import networkx as nx
G = nx.Graph()
```

```python
with open("day_23_input_0.txt") as fname:
    links = fname.read().splitlines()
links
```

```python
G.clear()

for link in links:
    computer_one, computer_two = link.split("-")
    G.add_edge(computer_one, computer_two)
```

```python
nx.draw_spring(G, with_labels=True, font_color="k", node_size=250, node_color="w")
```

```python
sorted(nx.simple_cycles(G, 3))
```

```python
triads_list = sorted(nx.simple_cycles(G, 3))
```

```python
accumulator = 0
for triad in triads_list:
    for computer in triad:
        if computer.startswith("t"):
            accumulator += 1
            break  # don't count more than once if several computers in a triad start with "t"
accumulator
```

```python
with open("day_23_input_1.txt") as fname:
    links = fname.read().splitlines()
```

```python
G.clear()

for link in links:
    computer_one, computer_two = link.split("-")
    G.add_edge(computer_one, computer_two)
```

```python
nx.draw_spring(G, with_labels=False, font_color="k", node_size=20, node_color="k")
```

```python
nx.write_graphml_lxml(G, "lanparty.graphml")
```

```python
triads_list = sorted(nx.simple_cycles(G, 3))
```

```python
accumulator = 0
for triad in triads_list:
    for computer in triad:
        if computer.startswith("t"):
            accumulator += 1
            break  # don't count more than once if several computers in a triad start with "t"
accumulator
```

## Part 2 example

```python
with open("day_23_input_0.txt") as fname:
    links = fname.read().splitlines()

G.clear()

for link in links:
    computer_one, computer_two = link.split("-")
    G.add_edge(computer_one, computer_two)
```

```python
cliques = list(nx.find_cliques(G))
cliques
```

```python
max(cliques, key=len)
```

```python
",".join(sorted(max(cliques, key=len)))
```
