## Example

```python
with open("day_18_input_0.txt") as fname:
    corruptedlocations = fname.read().splitlines()
```

```python
corruptedlocations
```

```python
import networkx as nx
G = nx.Graph()
G.clear()
```

```python
gridsize = 7
for row in range(gridsize):
    for col in range(gridsize):
        if row != 0:  # if not the very top row, make a connection between current node and the one above
            up_row = row - 1
            G.add_edge((row, col), (up_row, col))
        if row != (gridsize - 1):  # if not the very bottom row, make a connection between current node and the one below
            down_row = row + 1
            G.add_edge((row, col), (down_row, col))
        if col != 0:  # if not the leftmost column, make a connection between current node and the one to the left
            left_col = col - 1
            G.add_edge((row, col), (row, left_col))
        if col != gridsize - 1:  # if not the rightmost column, make a connection between current node and the one to the right
            right_col = col + 1
            G.add_edge((row, col), (row, right_col))

nx.draw_spring(G, with_labels=True, font_color="k", node_size=20, node_color="k")
```

```python
bytes_to_fall = 12

for node in range(bytes_to_fall):
    col, row = [int(x) for x in corruptedlocations[node].split(',')]
    G.remove_node((row, col))

nx.draw_spring(G, with_labels=True, font_color="k", node_size=20, node_color="k")
```

```python
start_pos = (0,0)
end_pos = (gridsize-1, gridsize-1)
nx.shortest_path_length(G, start_pos, end_pos)
```

```python
for row in range(gridsize):
    for col in range(gridsize):
        if row != 0:  # if not the very top row, make a connection between current node and the one above
            up_row = row - 1
            G.add_edge((row, col), (up_row, col))
        if row != (gridsize - 1):  # if not the very bottom row, make a connection between current node and the one below
            down_row = row + 1
            G.add_edge((row, col), (down_row, col))
        if col != 0:  # if not the leftmost column, make a connection between current node and the one to the left
            left_col = col - 1
            G.add_edge((row, col), (row, left_col))
        if col != gridsize - 1:  # if not the rightmost column, make a connection between current node and the one to the right
            right_col = col + 1
            G.add_edge((row, col), (row, right_col))

fallen_bytes = 0
while True:
    try:
        col, row = [int(x) for x in corruptedlocations[fallen_bytes].split(',')]
        G.remove_node((row, col))
        nx.shortest_path_length(G, start_pos, end_pos)
        fallen_bytes += 1
    
    except NetworkXNoPath:
        print(col,row, fallen_bytes)
        break
```
